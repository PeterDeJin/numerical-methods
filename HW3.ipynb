$$

#以下是用 Power method 的程式碼

```python
import numpy as np
import time

def infinite_norm(v):
    return np.max(np.abs(v))

def power_method(A, x, max_iter=1000, tol=1e-10):
    iteration_count = 0  # 迭代次數計數器
    for a in range(max_iter):
        iteration_count += 1
        x_next = np.dot(A, x)
        norm = infinite_norm(x_next)
        x_next = x_next / norm
        if infinite_norm(x_next - x) < tol:
            break
        x = x_next
        a += 1
    eigenvalue = np.dot(x.T, np.dot(A, x)) / np.dot(x.T, x)
    return eigenvalue, x, iteration_count

# 矩陣 A
A = np.array([[4, 1, 1, 1],
              [1, 3, 1, 1],
              [1, 1, 2, 0],
              [1, 1, 0, 2]])

# 初始向量 x
x0 = np.array([1, -2, 0, 3])

# 記錄開始時間
start_time = time.time()

eigenvalue, eigenvector, iteration_count = power_method(A, x0)

end_time = time.time()

# 輸出結果
print("eigenvalue : ", eigenvalue)
print("eigenvector : ", eigenvector)
print("迭代次數:", iteration_count)
print("耗時: {:.6f} 秒".format(end_time - start_time))
```

#接下來我想用 Aitken’s $Δ_{2}$ 去做加速

以下是有用 Aitken’s $Δ_{2}$ 加速的 Power method 程式碼

```python
import numpy as np
import time

def infinite_norm(v):
    return np.max(np.abs(v))

def aitken_delta_squared(seq):
    x_n, x_n1, x_n2 = seq
    return x_n - (x_n1 - x_n)**2 / (x_n2 - 2*x_n1 + x_n)

def power_method_with_aitken(A, x, max_iter=1000, tol=1e-10):
    seq = []
    iteration_count = 0  # 迭代次數計數器
    for a in range(max_iter):
        iteration_count += 1
        x = np.dot(A, x)
        x = x / infinite_norm(x)
        lambda_approx = np.dot(x.T, np.dot(A, x)) / np.dot(x.T, x)
        seq.append(lambda_approx)
        a += 1

        if len(seq) >= 3:
            lambda_aitken = aitken_delta_squared(seq[-3:])
            if abs(lambda_aitken - seq[-1]) < tol:
                return lambda_aitken, x, iteration_count

    return seq[-1], x, iteration_count

# 矩陣 A
A = np.array([[4, 1, 1, 1],
              [1, 3, 1, 1],
              [1, 1, 2, 0],
              [1, 1, 0, 2]])

# 初始向量 x
x0 = np.array([1, -2, 0, 3])

# 記錄開始時間
start_time = time.time()

# Aitken 的 Δ² 加速
eigenvalue, eigenvector, iteration_count = power_method_with_aitken(A, x0)

end_time = time.time()

# 輸出結果
print("eigenvalue : ", eigenvalue)
print("eigenvector : ", eigenvector)
print("迭代次數:", iteration_count)
print("耗時: {:.6f} 秒".format(end_time - start_time))
```

## 以下是用 Weilandt Deflation 並且用 $ L_{\infty} $ 算出第二個 eigenvalue 和 eigenvector

```python
import numpy as np

def infinite_norm(v):
    return np.max(np.abs(v))

def power_method(A, x, max_iter=1000, tol=1e-10):
    for a in range(max_iter):
        x_next = np.dot(A, x)
        x_next = x_next / infinite_norm(x_next)
        if infinite_norm(x_next - x) < tol:
            break
        x = x_next
        a += 1
    eigenvalue = np.dot(x.T, np.dot(A, x)) / np.dot(x.T, x)
    return eigenvalue, x

def deflation(A, x):
    lambda_1, u_1 = power_method(A, x)
    B = A - lambda_1 * np.outer(u_1, u_1) # B = A - lambda_1 * v * x^t
    lambda_2, u_2 = power_method(B, x)
    return lambda_2, u_2

# 矩陣 A
A = np.array([[4, 1, 1, 1],
              [1, 3, 1, 1],
              [1, 1, 2, 0],
              [1, 1, 0, 2]])

# 初始向量 x
x0 = np.array([1, -2, 0, 3])

second_eigenvalue, second_eigenvector = deflation(A, x0)

# 輸出结果
print("second eigenvalue : ", second_eigenvalue)
print("second eigenvector : ", second_eigenvector)
```
