Problem\
Use the conjugate gradient method to solve the linear system $ Ax=b $, where the entries of $ A $ and $ b $ are

$$
a_{i,j} =
\begin{cases}
2i, &\text{ when } j=i \text{ and } i = 1,2,....40,\\
-1, &\text{ when}
\begin{cases}
j = i + 1 \text{ and } i = 1,3,...,39,\\
j = i - 1 \text{ and } i = 2,4,...,40,\\
\end{cases}\\
0, &\text{  otherwise,}\\
\end{cases}
$$

and $b_{i} = 1.5i-6 $ , for each  $  i=1,...,40 .$

* (Optional) Use preconditioning with $C^{-1} = D^{-1/2}$

#以下是作業的解答，但我們還需要做一些確認，確認我的程式碼是沒有問題的

```python
import numpy as np
import time

def conjugate_gradient(A, b, tol=1e-10, max_iter=1000):
    x = np.zeros_like(b, dtype=np.float64)
    r = b - np.dot(A, x)
    v = r.copy()
    rdot = np.dot(r, r)

    for i in range(max_iter):
        Av = np.dot(A, v)
        t = rdot / np.dot(v, Av)
        x += t * v
        r -= t * Av
        rsnew = np.dot(r, r)
        if np.sqrt(rsnew) < tol:
            break
        v = r + (rsnew / rdot) * v
        rdot = rsnew
    return x

# 定義矩陣大小
n = 40

# 建立對角線元素
diagonal = np.array([2 * i for i in range(1, n + 1)])
off_diagonal = -1 * np.ones(n - 1)

# 建立矩陣Ａ
A = np.diag(diagonal) + np.diag(off_diagonal, k=1) + np.diag(off_diagonal, k=-1)

# 給定向量b
b = np.array([1.5 * i - 6 for i in range(1, n + 1)])

# 記錄時間
start_time = time.time()

x = conjugate_gradient(A, b)

end_time = time.time()

print("矩陣Ａ是")
print(A)
print("向量b是")
print(b)

# 輸出解
print("x的解是")
print(x)

# 計算時間
print("Conjugate Gradient Method took {:.6f} seconds".format(end_time - start_time))

```

```python
import numpy as np
import time

def conjugate_gradient(A, b, tol=1e-10, max_iter=1000):
    x = np.zeros_like(b, dtype=np.float64)
    r = b - np.dot(A, x)
    v = r.copy()
    rdot = np.dot(r, r)

    for i in range(max_iter):
        Av = np.dot(A, v)
        t = rdot / np.dot(v, Av)
        x += t * v
        r -= t * Av
        rsnew = np.dot(r, r)
        if np.sqrt(rsnew) < tol:
            break
        v = r + (rsnew / rdot) * v
        rdot = rsnew
    return x

# 建立矩陣Ａ
A = np.array([[0.2, 0.1, 1, 1, 0],
              [0.1, 4, -1, 1, -1],
              [1, -1, 60, 0, -2],
              [1, 1, 0, 8, 4],
              [0, -1, -2, 4, 700]])
# 向量b
b = np.array([1, 2, 3, 4, 5])

# 使用preconditioning  C^{-1} = D^{-1/2}
diagonal = np.diag(A)
D_inv_sqrt = np.diag(1 / np.sqrt(diagonal))
M = np.dot(np.dot(D_inv_sqrt, A), D_inv_sqrt)
b_precond = np.dot(D_inv_sqrt, b)

# 記錄時間
start_time = time.time()

x_precond = conjugate_gradient(M, b_precond)

x = np.dot(D_inv_sqrt, x_precond)

end_time = time.time()

print("矩陣Ａ是")
print(A)
print("向量b是")
print(b)

# 輸出解
print("x的解是")
print(x)

# 計算時間
print("Preconditioned Conjugate Gradient Method took {:.6f} seconds".format(end_time - start_time))
```
