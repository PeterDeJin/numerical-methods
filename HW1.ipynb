Problem\
Use the SOR method to solve the linear system Ax = b, where the entries of
A and b are \
$$
a_{i,j} =
\begin{cases}
2i & \text{when } j=i \text{ and } i = 1,2,...,N \\
0.5i & \text{when }  
\begin{cases}
j=i+2 \text{ and } i = 1,2,....,N-2\\
j=i-2 \text{ and } i = 3,4,....,N\\
\end{cases}\\
0.25i & \text{when }
\begin{cases}
j=i+4 \text{ and } i = 1,2,....,N-4\\
j=i-4 \text{ and } i = 5,6,....,N\\
\end{cases}\\
0 & \text{,otherwise}
\end{cases}
$$
and b<sub>i</sub> = π ,for each i = 1,2,....,N

* Set N =160 (or larger) and run SOR for different $ω = 0.7,0.8,0.9,1.0,1.1,1.2,1.3,1.4 .$
* Set the tolerance TOL = 10^(-6) and using the norm $||x^{(k)}-x^{(k-1)}||_{∞}$ as the stopping criterion.
* Output a table that shows the number of iterations and $||x^{(k)}-x^{(k-1)}||_{∞}$ of final approximate solutions for different values of $ω$



# 以下是作業的解答，但我們還需要做一些確認，確認我的程式碼是沒有問題的

```python
import numpy as np

def construct_matrix_A(N):
    A = np.zeros((N, N))
    for i in range(N):
        A[i, i] = 2 * (i + 1)
        if i + 2 < N:
            A[i, i + 2] = 0.5 * (i + 1)
            A[i + 2, i] = 0.5 * (i + 3)
        if i + 4 < N:
            A[i, i + 4] = 0.25 * (i + 1)
            A[i + 4, i] = 0.25 * (i + 5)
    return A

def sor(A, b, omega, tol=10**(-6), max_iterations=100000):
    N = len(A)
    x = np.zeros(N)
    x_new = np.copy(x)

    for iteration in range(max_iterations):
        for i in range(N):
            sums = np.dot(A[i, :i], x_new[:i]) + np.dot(A[i, i + 1:], x[i + 1:])
            x_new[i] = (1 - omega) * x[i] + (omega / A[i, i]) * (b[i] - sums)

        # 計算infinite norm
        diff = x_new - x
        max_diff = max(abs(diff))

        if max_diff < tol:
            return x_new, iteration + 1, max_diff
        else:
            x = np.copy(x_new)

    return x, max_iterations, max_diff

N = 160
A = construct_matrix_A(N)
b = np.pi * np.ones(N)

omegas = [0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4]
results = []

for omega in omegas:
    solution, iterations, norm = sor(A, b, omega)
    results.append((omega, iterations, norm, solution))

# 輸出結果
for result in results:
    print(f"Omega: {result[0]}, Iterations: {result[1]}, Final Norm: {result[2]:.10f}, Solution : {result[3:]}")
```
